With iloc, we can pass in the -1 to get the last row- something we couldn't do with loc.

```python
# get the first 100th, and 1000th row using iloc.
print(df.iloc[[0,99,999]])
```

1.The python slicing syntax uses a colon,:
2.if we have just a colon,the attribute refers to everything.
3.So,if we just want to get the first column using the loc or iloc syntax , we can write something like df.loc[:,[columns]] to subset the colmns.

```python
subset = df.loc[:,['year','pop']]
print(subset.head())
```

```python
#subset columns with iloc
# iloc will allow us to use integers.
# -1 will select the last column.
subset = df.iloc[:,[2,4,-1]]
print(subset.head())
```

```python
# or using range to subset values
print(df.iloc[:,range(0,5)])
```

```python
# using loc
print(df.loc[42,'country'])
```

```python
# using iloc
print(df.iloc[42,0])
```

```python
# get the 1st ,100th,1000th rows
# from the 1st,4th and 6th columns
print(df.iloc[[0,99,999],[0,3,5]])
#using loc
print(df.loc[10:13,['country','lifeExp','gdpPercap']])
```

```python
# Grouped Means
1.# For each year in our
# data,what was the average life expectancy?
print(df.groupby('year')['lifeExp'].mean())
```

```python
multi_group_var = df.groupby(['year','continent'])\
[['lifeExp','gdpPercap']].mean()
print(multi_group_var)
```

```python
# If you need to flatten the dataframe,you can use the 
# reset_index method.
flat= multi_group_var.reset_index()
print(flat.head(15))
```

```python
# Grouped Frequency counts
# Use the nunique to get counts of unique values on a Pandas Series.
print(df.groupby('continent')['country'].nunique())
```

```python
# Basic Plot
global_yearly_life_expectancy = df.groupby('year')['lifeExp'].mean()
print(global_yearly_life_expectancy)
```
